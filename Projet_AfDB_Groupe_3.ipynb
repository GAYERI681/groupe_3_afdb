```python
# Fonction pour reformater les dates
def reformat_date(data_Date_str):
    if isinstance(data_Date_str, str) and 'Q' in data_Date_str:  # Check if it's a string first
        year, quarter = data_Date_str.split('Q')
        month = int(quarter) * 3
        return f"{year}-{month:02}-01"
    elif isinstance(data_Date_str, str) and 'M' in data_Date_str:  # Check if it's a string first
        year, month = data_Date_str.split('M')
        return f"{year}-{month:02}-01"
    else:
        return f"{data_Date_str}-01-01"

df_Data['Formatted data_Date'] = df_Data['Date'].apply(reformat_date)
df_Data
```

```python
# Créer un dictionnaire pour stocker les totaux par région
region_totals = {}

# Itérer sur les régions
for region in ['North_Africa', 'East_Africa', 'Central_Africa', 'West_Africa', 'Southern_Africa']:
    # Créer une liste pour stocker les valeurs pour chaque pays dans la région
    country_values = []
    # Itérer sur les lignes du DataFrame
    for index, row in df_Data.iterrows():
        # Si le pays est dans la région actuelle, ajouter sa valeur à la liste
        if row[region] != '':
            country_values.append(row['operationsValue'])
    # Calculer la somme des valeurs pour la région et l'ajouter au dictionnaire
    region_totals[region] = sum(country_values)

# Créer un DataFrame à partir du dictionnaire
df_region_totals = pd.DataFrame.from_dict(region_totals, orient='index', columns=['Total Operations Value'])

# Afficher le DataFrame
print(df_region_totals)
```

```python
# Créer un dictionnaire pour stocker les totaux par région et par pays
region_country_totals = {}

# Itérer sur les régions
for region in ['North_Africa', 'East_Africa', 'Central_Africa', 'West_Africa', 'Southern_Africa']:
  region_country_totals[region] = {}
  # Itérer sur les pays dans la région
  for country in df_Data[region].unique():
    if country != '':  # Ignorer les valeurs vides
      # Calculer le total des operationsValue pour le pays dans la région
      total = df_Data.loc[(df_Data[region] == country), 'operationsValue'].sum()
      region_country_totals[region][country] = total

# Créer un DataFrame à partir du dictionnaire
df_region_country = pd.DataFrame.from_dict(region_country_totals, orient='index').stack().reset_index()
df_region_country.columns = ['Region', 'Country', 'Total Operations Value']

# Afficher le DataFrame
print(df_region_country)
```
